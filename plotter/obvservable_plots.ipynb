```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

def count_live_cells_in_frame(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return sum(1 for cell in data if cell["state"] == "ALIVE")

def normalize_runs(runs_data, num_points=100):
    max_length = num_points
    normalized_runs = []
    
    for run in runs_data:
        normalized_run = np.interp(np.linspace(0, len(run) - 1, max_length),
                                   np.arange(len(run)),
                                   run)
        normalized_runs.append(normalized_run)
    
    normalized_runs = np.array(normalized_runs)
    mean_values = np.mean(normalized_runs, axis=0)
    std_dev = np.std(normalized_runs, axis=0)
    
    return mean_values, std_dev

def process_directory(base_path, target_density, num_points=100):
    systems_data = {}
    for system in os.listdir(base_path):
        system_path = os.path.join(base_path, system)
        if os.path.isdir(system_path):
            density_path = os.path.join(system_path, target_density)
            if os.path.isdir(density_path):
                runs_data = []
                for run in os.listdir(density_path):
                    run_path = os.path.join(density_path, run)
                    if os.path.isdir(run_path):
                        frames = []
                        for frame_file in sorted(os.listdir(run_path)):
                            if frame_file.startswith("frame") and frame_file.endswith(".json"):
                                frame_path = os.path.join(run_path, frame_file)
                                frames.append(count_live_cells_in_frame(frame_path))
                        if frames:
                            runs_data.append(frames)
                if runs_data:
                    systems_data[system] = normalize_runs(runs_data, num_points)
    return systems_data

def plot_results(systems_data, target_density):
    plt.figure(figsize=(12, 8))
    for system, (mean_values, std_dev) in systems_data.items():
        plt.plot(mean_values, label=f'{system}')
        plt.fill_between(range(len(mean_values)), mean_values - std_dev, mean_values + std_dev, alpha=0.2)
    plt.xlabel('Normalized Frame')
    plt.ylabel('Average Live Cells')
    plt.title(f'Average Live Cells per System (Density {target_density})')
    plt.ylim(bottom=0)  # Fix the y-axis origin at 0.0
    plt.xlim(0, 100)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Move legend outside the plot
    plt.grid(True)
    plt.tight_layout()  # Adjust layout to make room for the legend
    plt.show()

if __name__ == "__main__":
    base_path = "../GOF/output"  # Cambia esto por la ruta real del directorio base
    target_density = "0.2"  # Cambia esto a la densidad que quieres graficar
    num_points = 100
    systems_data = process_directory(base_path, target_density, num_points)
    plot_results(systems_data, target_density)
```
